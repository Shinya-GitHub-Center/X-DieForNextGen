以上を踏まえた上で、私が考える「身なりと賃金の相関関係」のヒートグラフを示す。  
濃い赤色になるほど「その通り！」の関係が強くなり、濃い青色になるほど「違います！」の関係が強くなります。

```python
import seaborn as sns
import pandas as pd

sns.set(font='HackGen Console')
list = [[-0.25, 0.25], [0.25, -0.25]]
df = pd.DataFrame(data=list, index=['Tシャツ＆サンダル姿', 'きちんとした格好'], columns=['高収入', '低収入'])
sns.heatmap(df, cmap="jet", vmin=-1.0, vmax=1.0)
```
